```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("public_services_dataset.xlsx")
df.head()
df.info()
df.columns
df.describe()



daily_system = df.groupby("OCCUPANCY_DATE")["OCCUPANCY_RATE"].mean()

plt.figure()
daily_system.plot()
plt.title("System-Wide Average Occupancy Rate")
plt.ylabel("Occupancy Rate")
plt.show()
```
